```python
# prepare directories
import os
if not os.path.isdir('mov'):
  os.mkdir('mov')
if not os.path.isdir('mp4'):
  os.mkdir('mp4')
!unzip '/content/drive/MyDrive/Sharing is Caring/Academic/Digital Video Project/data/landscapes.zip' -d mov
!pip install ffmpeg
```

```python
# conversion
mov_dir = 'mov'
mov_files = sorted(os.listdir(mov_dir))
for i, input in enumerate(mov_files):
  if '.mov' in input:
    input = os.path.join(mov_dir, input)
    output = os.path.join('mp4', f'{i}.mp4')
    !ffmpeg -i $input $output
!zip -r '/content/drive/MyDrive/Sharing is Caring/Academic/Digital Video Project/data/landscapes.zip' mp4
```
